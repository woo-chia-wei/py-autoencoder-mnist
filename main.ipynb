```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping
from skimage import color
from skimage import io
```

```python
X = np.load('X_MNIST.npy')
y = np.load('y_MNIST.npy')
X_noise = np.load('X_MNIST_noise.npy')
```

```python
autoencoder = Sequential()

autoencoder.add(Dense(128, input_shape=(784, ), activation="relu"))
autoencoder.add(Dense(64, activation="relu"))
autoencoder.add(Dense(32, activation="relu"))
autoencoder.add(Dense(64, activation="relu"))
autoencoder.add(Dense(128, activation="relu"))
autoencoder.add(Dense(784, activation="sigmoid"))
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

history = autoencoder.fit(X, X, epochs=1000, verbose=1, callbacks=[EarlyStopping(patience=3)])
```

# Original vs Reconstructed Images

```python
nrows=2
ncols=8

fig, axes = plt.subplots(figsize=(2 * ncols, 2 * nrows), nrows=nrows, ncols=ncols)
axes = axes.flatten()

for ax, image in zip(axes, X):
    ax.imshow(image.reshape((28, 28)), interpolation='nearest', cmap='gray')
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()

fig, axes = plt.subplots(figsize=(2 * ncols, 2 * nrows), nrows=nrows, ncols=ncols)
axes = axes.flatten()

for ax, image in zip(axes, autoencoder.predict(X)):
    ax.imshow(image.reshape((28, 28)), interpolation='nearest', cmap='gray')
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
```

# Noise vs Reconstructed Images

```python
nrows=2
ncols=8

fig, axes = plt.subplots(figsize=(2 * ncols, 2 * nrows), nrows=nrows, ncols=ncols)
axes = axes.flatten()

for ax, image in zip(axes, X_noise):
    ax.imshow(image.reshape((28, 28)), interpolation='nearest', cmap='gray')
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()

fig, axes = plt.subplots(figsize=(2 * ncols, 2 * nrows), nrows=nrows, ncols=ncols)
axes = axes.flatten()

for ax, image in zip(axes, autoencoder.predict(X_noise)):
    ax.imshow(image.reshape((28, 28)), interpolation='nearest', cmap='gray')
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
```

# Handwritten vs Reconstructed Images

```python
def load_written_images(filenames):
    return np.asarray([cv2.resize(color.rgb2gray(io.imread(filename)), 
                      (28, 28), interpolation = cv2.INTER_AREA).reshape(784) for filename in filenames])

written_images = load_written_images([f'written_{digit}.png' for digit in range(10)])

nrows=2
ncols=5

fig, axes = plt.subplots(figsize=(2 * ncols, 2 * nrows), nrows=nrows, ncols=ncols)
axes = axes.flatten()

for ax, image in zip(axes, written_images):
    ax.imshow(image.reshape(28, 28), interpolation='nearest', cmap='gray')
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()

fig, axes = plt.subplots(figsize=(2 * ncols, 2 * nrows), nrows=nrows, ncols=ncols)
axes = axes.flatten()

for ax, image in zip(axes, autoencoder.predict(written_images)):
    ax.imshow(image.reshape(28, 28), interpolation='nearest', cmap='gray')
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
```
